```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow import keras
import numpy as np
```

```python
(X_train, Y_train), (X_test, Y_test) = keras.datasets.mnist.load_data()
```

```python
print(f"size of train set is {len(X_train)} images and size of test set is {len(X_test)} images")
print(f"size of each image is {X_train[0].shape}")
```

```python
#checking one sample image
plt.matshow(X_train[15])
```

```python
#seeing the output Y
print(f"{Y_train[15]} is the sample output for the above image")
```

```python
# Flattening the data to make it ready for the CNN
print(f"current shape of the dataset is {X_train.shape}")
#making it to 60000*784
Xtrain_flat = X_train.reshape(len(X_train), X_train[0].shape[0]*X_train[0].shape[1])
print(f"after flattening is {Xtrain_flat.shape}")
```

```python
#doing for X_test as well
Xtest_flat = X_test.reshape(len(X_test), X_test[0].shape[0]*X_test[0].shape[1])
print(f"after flattening is {Xtest_flat.shape}")
```

```python
#initiating a bare run

model = keras.Sequential([
    keras.layers.Dense( 10, input_shape = (784, ), activation ="sigmoid")
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(Xtrain_flat, Y_train, epochs = 5)
```

```python
#scaling our data
X_train = X_train/255
Xtrain_flat = X_train.reshape(len(X_train), X_train[0].shape[0]*X_train[0].shape[1])
X_test = X_test/255
Xtest_flat = X_test.reshape(len(X_test), X_test[0].shape[0]*X_test[0].shape[1])
```

```python
#now checking the accuracy

model = keras.Sequential([
    keras.layers.Dense( 10, input_shape = (784, ), activation ="sigmoid")
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(Xtrain_flat, Y_train, epochs = 5)
```

```python
model.evaluate(Xtest_flat, Y_test)
```

```python
y_predicted = model.predict(Xtest_flat)
y_prediction = [np.argmax(i) for i in y_predicted]
print(f'5th image prediction is {y_prediction[5]}')
```

```python
#checking for a single prediction
plt.matshow(X_test[5])
```

```python
#drawing Confusion Matrix

cm = tf.math.confusion_matrix(labels = Y_test, predictions = y_prediction)

import seaborn as sb
plt.figure(figsize = (10,7))
sb.heatmap(cm , annot = True, fmt ='d')
plt.xlabel = "predicted"
plt.ylabel = "truth"
```
